**Questão 1 - d)** Como estão distribuídas as avaliações ao longo do tempo?
<br/><br/>


```python
import matplotlib.pyplot as plt
```

```python
# Convertendo a coluna timestamp para datetime
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s')
```

```python
# Extraindo o ano
ratings['year'] = ratings['timestamp'].dt.year
```

```python
# Agrupando por ano e conta avaliações
ratings_per_year = ratings.groupby('year')['rating'].count()
```

```python
# Plotando o gráfico
plt.figure(figsize=(10,6))
ratings_per_year.plot(kind='bar', color='skyblue')
plt.title('Distribuição das Avaliações ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Avaliações')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**Questão 2 - Movies - b)** Existem quantos filmes de Aventura, quantos de Romance, e quantos com os dois juntos?

```python
filtro = movies['Generos'].str.contains('Adventure')
filtro2 = movies['Generos'].str.contains('Romance')
filtro3 = movies['Generos'].str.contains(r'(?=.*Adventure)(?=.*Romance)', regex=True)

adventureMovies = movies[filtro]
romanceMovies = movies[filtro2]
adventureRomanceMovies = movies[filtro3]
```
